```python
import folium
import urllib2
import json
import webbrowser

# GeoJSON url feed from USGS website: http://earthquake.usgs.gov/earthquakes/feed/v1.0/geojson.php
url_json = "http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson"

# HTTP request
req = urllib2.Request(url_json)

# Open the URL and read the data
webUrl = urllib2.urlopen(req)
data = webUrl.read()

# Decoding JSON
theJSON = json.loads(data)


if "title" in theJSON["metadata"]:
    # Displaying title
    print theJSON["metadata"]["title"]    
    # Displaying recorded earthquake events    
    count = theJSON["metadata"]["count"]
    print "{0} events recorded.".format(str(count))

# Displaying Map
map = folium.Map(location=[20, -102], zoom_start=2)

# Getting coordinates, location & earthquake magnitude on to map
for i in theJSON['features']:
    # Getting longitude and latitude coordinates
    coordinates = i['geometry']['coordinates'][:2][::-1]
    # Getting location name
    locate = i['properties']['place'].encode("utf-8")
    # Getting earthquake magnitude
    magnitude = i['properties']['mag']
    str_mag = str(magnitude)
    # Displaying location markers on map
    map.simple_marker(location=coordinates, popup=locate + " and the magnitude of the earth quake is {}".format(str_mag))

# Saving map to html
map.create_map(path='map.html')

# Displaying map on browser tab
webbrowser.open_new_tab('map.html')

# Displaying map on Jupyter Notebook
map
```
